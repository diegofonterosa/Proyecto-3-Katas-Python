## EJERCICIO 1
### Escribe una función que reciba una cadena de texto como parámetro
### y devuelva un diccionario con las frecuencias de cada letra en la cadena.
### Los espacios no deben ser considerados.

```python
def frecuencia_letras(texto):
    #Frecuencia de cada letra sin contar espacios
    frecuencias = {}
    
    for caracter in texto:
        #Ignorar los espacios
        if caracter != ' ':
            if caracter in frecuencias:
                #Si hay más de una letra igual se incrementa su número
                frecuencias[caracter] += 1
            else:
                #Si es la primera vez que aparece empieza a contar en 1
                frecuencias[caracter] = 1
    return frecuencias

texto_prueba = "Hola mundo"
print("Ejercicio 1:", frecuencia_letras(texto_prueba))
```

### como parámetros. La función debe devolver una lista con todas las palabras

```python
def buscar_palabra_en_lista(lista_palabras, palabra_objetivo):
    # Busca palabras que contengan la palabra objetivo
    palabras_encontradas = []
    
    for palabra in lista_palabras:
        if palabra_objetivo in palabra:
            palabras_encontradas.append(palabra)
            
    return palabras_encontradas

# Prueba
palabras = ["python", "progrmacion", "pythonista", "java", "desarrollo"]
objetivo = "python"
print("Ejercicio 3:", buscar_palabra_en_lista(palabras, objetivo))
```

## EJERCICIO 5
### Escribe una función que tome una lista de números como parámetro 
### y un valor opcional nota_aprobado (por defecto 5).
### La función debe calcular la media de los números de la lista y
### determinar si la media es mayor o igual que nota_aprobado.

```python
def calcular_media_y_estado(notas, nota_aprobado=5):
    # Calcula media y determina si está aprobado o suspenso. Devuelve una tupla
    
    media = sum(notas) / len(notas)
    
    if media >= nota_aprobado:
        estado = "aprobado"
    else:
        estado = "suspenso"
        
    return (media, estado)

# Prueba
notas_alumno = [6, 4, 5, 3, 4]
print("Ejercicio 5:", calcular_media_y_estado(notas_alumno))
```

## EJERCICIO 6
### Escribe una función que calcule el factorial de un número
### de manera recursiva

```python
def factorial(n):
    # Calcula el factorial de manera recursiva
    if n == 0 or n == 1:
        return 1
    else :
        return n * factorial(n - 1)
    
# Prueba
numero = 5
print(f"Ejercicio 6: el factorial de {numero} es {factorial(numero)}")
```

## EJERCICIO 10
### Escribe una función que reciba una lista de números y calcule su promedio.
### Si la lista está vacia, lanza una excepción personalizada.

```python
class ListaVaciaError(Exception):
    # Excepción para la lista vacia
    pass

def calcular_promedio(numeros):
    # Calcula el promedio de una lista de números
    
    try:
        if len(numeros) == 0:
            raise ListaVaciaError("Error: la lista está vacía, no se puede calcular el promedio")
        
        promedio = sum(numeros) / len(numeros)
        return promedio
    
    except ListaVaciaError as e:
        return str(e)
    
# Prueba
print("Ejercicio 10:")
print("Promedio de [5, 10, 15]:", calcular_promedio([5, 10, 15]))
print("Promedio de []:", calcular_promedio([]))
```

## EJERCICIO 28
### Crea una función que busque y devuelva el primer elemento
### duplicado en una lista dada.

```python
def primer_duplicado(lista):
    # Encuentra el primer elemento duplicado en una lista
    
    vistos = set()
    
    for elemento in lista:
        if elemento in vistos:
            return elemento
        vistos.add(elemento)
        
    return None

# Prueba
lista = [1, 2, 3, 4, 2, 5, 6]
print("Ejercicio 28:", primer_duplicado(lista))
```

## EJERCICIO 29
### Crea una función que convierta una variable en una cadena de texto
### y enmascare todos los caracteres con el carácter '#' excepto los
### últimos cuatro.

```python
def enmascarar_texto(variable):
    # Enmascara todos los caracteres excepto los últimos 4
    
    texto = str(variable)
    
    if len(texto) <= 4:
        return texto
    else:
        parte_enmascarada = '#' * (len(texto) - 4)
        ultimos_cuatro = texto[-4:]
        return parte_enmascarada + ultimos_cuatro
    
# Prueba
numero_tarjeta = "1234567890123456"
print("Ejercicio 29:", enmascarar_texto(numero_tarjeta))
```

```python
def son_anagramas(palabra1, palabra2):
    # Verifica si dos palabras son anagramas
    
    palabra1_ordenada = sorted(palabra1.lower())
    palabra2_ordenada = sorted(palabra2.lower())
    
    return palabra1_ordenada == palabra2_ordenada

# Prueba
print("Ejercicio 30:")
print("'amor' y 'roma'", son_anagramas("amor", "roma"))
print("'hola' y 'adios'"), son_anagramas("hola", "adios")
```

## EJERCICIO 34
### Crea la calse Arbol

```python
class Arbol:
    # Clase que representa un árbol genérico con tronco y ramas
    
    def __init__(self):
        self.tronco = 1
        self.ramas = []
        
    def crecer_tronco(self):
        self.tronco += 1
        
    def nueva_rama(self):
        self.ramas.append(1)
        
    def crecer_ramas(self):
        self.ramas = [rama + 1 for rama in self.ramas]
        
    def quitar_rama(self, posicion):
        if 0 <= posicion < len(self.ramas):
            self.ramas.pop(posicion)
        else:
            print("Posición inválida")
            
    def info_arbol(self):
        info = f"Longitud del tronco: {self.tronco}\n"
        info += f"Número de ramas: {len(self.ramas)}\n"
        info += f"Longitudes de las ramas: {self.ramas}"
        return info
    
# Caso de uso
print("\nEjercicio 34 - Caso de uso de la clase Arbol:")
arbol = Arbol()
print("a. Árbol creado")

arbol.crecer_tronco()
print("b. Tronco ha crecico")

arbol.nueva_rama()
print("c. Nueva rama añadida")

arbol.crecer_ramas()
print("d. Todas la ramas han crecido")

arbol.nueva_rama()
arbol.nueva_rama()
print("e. Dos nuevas ramas añadidas")

arbol.quitar_rama(2)
print("f. Rama en posición 2 retirada")

print("g. Información del árbol:")
print(arbol.info_arbol())
```

## EJERCICIO 35
### Crea la clase UsuarioBanco

```python
class UsuarioBanco:
    # Clase que representa un usuario de banco
    
    def __init__(self, nombre, saldo, tiene_cuenta_corriente):
        # Inicializa usuario con nombre, saldo y tipo de cuenta
        self.nombre = nombre
        self.saldo = saldo
        self.tiene_cuenta_corriente = tiene_cuenta_corriente
        
    def retirar_dinero(self, cantidad):
        
        if cantidad > self.saldo:
            raise ValueError(f"Error: Saldo insuficiente. Saldo actual: {self.saldo}")
        
        self.saldo -= cantidad
        print(f"{self.nombre} ha retirado {cantidad}. Saldo actual: {self.saldo}")
        
    def transferir_dinero(self, otro_usuario, cantidad):
        
        if cantidad > self.saldo:
            raise ValueError(f"Error: Saldo insuficiente para transferir {cantidad}")
        
        self.saldo -= cantidad
        otro_usuario.saldo += cantidad
        print(f"Tranferencia de {cantidad} de {self.nombre} a {otro_usuario.nombre}")
        
    def agregar_dinero(self, cantidad):
        
        self.saldo += cantidad
        print(f"{self.nombre} ha agregado {cantidad}. Saldo actual: {self.saldo}")
        
    def __str__(self):
        
        return f"{self.nombre} - Saldo: {self.saldo}"
    
# Caso de uso
print("\nEjercicio 35 - Caso de uso de la clase UsuarioBanco:")
alicia = UsuarioBanco("Alicia", 100, True)
bob = UsuarioBanco("Bob", 50, True)

print("a. Usuarios creados:")
print(alicia)
print(bob)

print("\nb. Agregar 20 a Bob:")
bob.agregar_dinero(20)

print("\nc. Transferir 80 de Bob a Alicia:")
try:
    bob.transferir_dinero(alicia, 80)
    print(f"Saldo Bob: {bob.saldo}, Saldo Alicia: {alicia.saldo}")
except ValueError as e:
    print(e)
    
print("\nd. Retirar 50 de Alicia:")
try:
    alicia.retirar_dinero(50)
except ValueError as e:
    print(e)    
```

## EJERICIO 36
### Crea una función llamada procesar_texto

```python
def contar_palabras(texto):
    # Cuenta cuantas veces aparece cada palabra en el texto
    
    palabras = texto.lower().split()
    frecuencias = {}
    
    for palabra in palabras:
        if palabra in frecuencias:
            frecuencias[palabra] += 1
        else:
            frecuencias[palabra] = 1
    
    return frecuencias

def reemplazar_palabras(texto, palabra_original, palabra_nueva):
    
    return texto.replace(palabra_original, palabra_nueva)

def eliminar_palabra(texto, palabra):
    
    palabras = texto.split()
    palabras_filtradas = [p for p in palabras if p != palabra]
    return ' '.join(palabras_filtradas)

def procesar_texto(texto, opcion, *args):
    
    if opcion == "contar":
        return contar_palabras(texto)
    elif opcion == "reemplazar":
        palabra_original, palabra_nueva = args
        return reemplazar_palabras(texto, palabra_original, palabra_nueva)
    elif opcion == "eliminar":
        palabra = args[0]
        return eliminar_palabra(texto, palabra)
    else:
        return "Opción no válida"
    
# Caso de uso
print("\nEjericio 36 - Caso de uso de procesar_texto:")
texto_ejemplo = "Python es genial Python es poderoso"

print("Opción contar:")
print(procesar_texto(texto_ejemplo, "contar"))

print("\nOpción reemplazar:")
print(procesar_texto(texto_ejemplo, "reemplazar", "Python", "JavaScript"))

print("\nOpción eliminar:")
print(procesar_texto(texto_ejemplo, "eliminar", "es"))
```

### Genera un programa que indique si es de noche, de día o de tarde

```python
def determinar_periodo_dia(hora):
    # Determina si es noche, día o tarde según la hora
    
    if 6 <= hora < 12:
        return "Es de día"
    elif 12 <= hora < 20:
        return "Es de tarde"
    else:
        return "Es de noche"
    
# Prueba
print("\nEjericio 37:")
print(f"Hora 9: {determinar_periodo_dia(9)}")
print(f"Hora 15: {determinar_periodo_dia(15)}")
print(f"Hora 22: {determinar_periodo_dia(22)}")
```

## EJERCICIO 38
### Escribe un programa que dtermine qué calificación en texto
### tiene un alumno según su calificación numérica.

```python
def obtener_calificacion_texto(nota):
    # Convierte calificación numérica a texto
    
    if 0 <= nota <= 69:
        return "Insuficiente"
    elif 70 <= nota <= 79:
        return "Bien"
    elif 80 <= nota <= 89:
        return "Muy bien"
    elif 90 <= nota <= 100:
        return "Excelente"
    else:
        return "Nota inválida"
    
# Prueba
print(f"Nota 105: {obtener_calificacion_texto(105)}")
print(f"Nota 75: {obtener_calificacion_texto(75)}")
print(f"Nota 85: {obtener_calificacion_texto(85)}")
print(f"Nota 95: {obtener_calificacion_texto(95)}")
```

## EJERCICIO 39
### Escribe una función que calcule el área de diferentes figuras geométricas

```python
import math

def calcular_area(figura, datos):
    # Calcula el área de una figura geométrica
    
    if figura == "rectangulo":
        base, altura = datos
        area = base * altura
        return f"Área del rectángulo: {area}"
    
    elif figura == "circulo":
        radio = datos[0]
        area = math.pi * (radio ** 2)
        return f"Área del círculo: {area:.2f}"
    
    elif figura == "triangulo":
        base, altura = datos
        area = (base * altura) / 2
        return f"Área del triángulo: {area}"
    else:
        return "Figura no reconocida"
    
# Prueba
print("\nEjericio 39:")
print(calcular_area("rectangulo", (5, 10)))
print(calcular_area("circulo", (7,)))
print(calcular_area("triangulo", (6, 8)))
```

## EJERCICIO 40
### Programa para determinar el monto final de una compra con descuento

```python
def calcular_precio_final():
    # Calcula el precio final depués de aplicar un descuento
    
    try:
        precio_original = float(input("Introduce el precio original del artículo: "))
        
        tiene_cupon = input("¿Tienes cupón de descuento? (si/no): ").lower()
        
        precio_final = precio_original
        
        if tiene_cupon == "si":
            
            valor_cupon = float(input("Introduce el valor del cupón de descuento (%): "))
            
            if valor_cupon > 0:
                descuento = precio_original * (valor_cupon / 100)
                precio_final = precio_original - descuento
                print(f"\nDescuento aplicado: {descuento:.2f}€")
            else:
                print("\nValor de cupón inválido. No se aplica descuento")
                
        elif tiene_cupon == "no":
            print("\nNo se aplicará descuento")
            
        else:
            print("\nRespuesta no válida. No se aplicará descuento")
            
        print(f"Precio final de la compra: {precio_final:.2f}€")
        
    except ValueError:
        print("Error: Debes introducir valores numéricos válidos")
        
```
