## EJERCICIO 1
### Escribe una función que reciba una cadena de texto como parámetro
### y devuelva un diccionario con las frecuencias de cada letra en la cadena.
### Los espacios no deben ser considerados.

```python
def frecuencia_letras(texto):
    #Frecuencia de cada letra sin contar espacios
    frecuencias = {}
    
    for caracter in texto:
        #Ignorar los espacios
        if caracter != ' ':
            if caracter in frecuencias:
                #Si hay más de una letra igual se incrementa su número
                frecuencias[caracter] += 1
            else:
                #Si es la primera vez que aparece empieza a contar en 1
                frecuencias[caracter] = 1
    return frecuencias

texto_prueba = "Hola mundo"
print("Ejercicio 1:", frecuencia_letras(texto_prueba))
```

## EJERCICIO 2
### Dada una lista de números, obtén una nueva lista con el doble
### de cada valor. Usa la función map().

```python
def duplicar_valores(lista):
    # Devuelve el doble de cada valor usando map()
    return list(map(lambda x: x * 2, lista))

# Prueba
numeros = [1, 2, 3, 4, 5]
print("Ejercicio 2:", duplicar_valores(numeros))
```

## EJERCICIO 3
### Escribe una función que tome una lista de palabras y una palabra objetivo
### como parámetros. La función debe devolver una lista con todas las palabras
### de la lista original que contengan la palabra objetivo.

```python
def buscar_palabra_en_lista(lista_palabras, palabra_objetivo):
    # Busca palabras que contengan la palabra objetivo
    palabras_encontradas = []
    
    for palabra in lista_palabras:
        if palabra_objetivo in palabra:
            palabras_encontradas.append(palabra)
            
    return palabras_encontradas

# Prueba
palabras = ["python", "progrmacion", "pythonista", "java", "desarrollo"]
objetivo = "python"
print("Ejercicio 3:", buscar_palabra_en_lista(palabras, objetivo))
```

## EJERCICIO 4
### Genera una función que calcule la diferencia entre los valores
### de dos listas. Usa la función map().

```python
def diferencia_listas(lista1, lista2):
    # Calcula la diferencia entre elementos de dos listas usando map()
    
    return list(map(lambda x, y: x - y, lista1, lista2))

# Prueba 
lista_a = [10, 20, 30, 40]
lista_b = [5, 10, 15, 20]
print("Ejercicio 4:", diferencia_listas(lista_a, lista_b))
```

## EJERCICIO 5
### Escribe una función que tome una lista de números como parámetro 
### y un valor opcional nota_aprobado (por defecto 5).
### La función debe calcular la media de los números de la lista y
### determinar si la media es mayor o igual que nota_aprobado.

```python
def calcular_media_y_estado(notas, nota_aprobado=5):
    # Calcula media y determina si está aprobado o suspenso. Devuelve una tupla
    
    media = sum(notas) / len(notas)
    
    if media >= nota_aprobado:
        estado = "aprobado"
    else:
        estado = "suspenso"
        
    return (media, estado)

# Prueba
notas_alumno = [6, 4, 5, 3, 4]
print("Ejercicio 5:", calcular_media_y_estado(notas_alumno))
```

## EJERCICIO 6
### Escribe una función que calcule el factorial de un número
### de manera recursiva

```python
def factorial(n):
    # Calcula el factorial de manera recursiva
    if n == 0 or n == 1:
        return 1
    else :
        return n * factorial(n - 1)
    
# Prueba
numero = 5
print(f"Ejercicio 6: el factorial de {numero} es {factorial(numero)}")
```

## EJERICIO 7
### Genera una función que convierta una lista de tuplas a una
### lista de strings. Usa la función map().

```python
def tuplas_a_strings(lista_tuplas):
    # Convierte una lista de tuplas a una lista de strings usando map()
    
    return list(map(lambda tupla: str(tupla), lista_tuplas))

# Prueba
tuplas = [(1, 2), (3, 4), (5, 6)]
print("Ejercicio 7:", tuplas_a_strings(tuplas))
```

##  EJERCICIO 8
### Escribe un programa que pida al usuario dos números e intente dividirlos.
### Si el usuario ingresa un valor no numérico o intenta dividir por cero, 
### maneja esas excepciones de manera adecuada

```python
def dividir_numeros():
# Pide dos números al usuario y los divide

    try:
        num1 = float(input("Introduce el primer número: "))
        num2 = float(input("Introduce el segundo número: "))
        
        resultado = num1 / num2
        print(f"El resultado de la división es: {resultado}")
        print("División exitosa")
        
    except ValueError:
        print("Error: Debes introducir valores numéricos")
    except ZeroDivisionError:
        print("Error: No se puede dividir por cero")
    except Exception as e:
        print(f"Error inesperado: {e}")
```

## EJERCICIO 9
### Escribe una función que tome una lista de nombres de mascotas
### como parámetro y devuelva una nueva lista excluyendo ciertas
### mascotas prohibidas en España. Usa la función filter

```python
def filtrar_mascotas_permitidas(lista_mascotas):
    # Filtra mascotas permitidas, excluyendo las prohibidas
    mascotas_prohibidas = ["Mapache", "Tigre", "Serpiente Pitón", "Cocodrilo", "Oso"]
    
    return list(filter(lambda mascota: mascota not in mascotas_prohibidas, lista_mascotas))

# Prueba
mascotas = ["Perro", "Gato", "Mapache", "Conejo", "Tigre", "Hamster", "Oso"]
print("Ejercicio 9:", filtrar_mascotas_permitidas(mascotas))
```

## EJERCICIO 10
### Escribe una función que reciba una lista de números y calcule su promedio.
### Si la lista está vacia, lanza una excepción personalizada.

```python
class ListaVaciaError(Exception):
    # Excepción para la lista vacia
    pass

def calcular_promedio(numeros):
    # Calcula el promedio de una lista de números
    
    try:
        if len(numeros) == 0:
            raise ListaVaciaError("Error: la lista está vacía, no se puede calcular el promedio")
        
        promedio = sum(numeros) / len(numeros)
        return promedio
    
    except ListaVaciaError as e:
        return str(e)
    
# Prueba
print("Ejercicio 10:")
print("Promedio de [5, 10, 15]:", calcular_promedio([5, 10, 15]))
print("Promedio de []:", calcular_promedio([]))
```

## EJERCICIO 11

### Si el usuario ingresa un valor no numérico o un valor fuera del
### rango esperado (menor que 0 o mayor que 120), maneja las excepciones.

```python
def validad_edad():
    # Pide edad al usuario y valida que sea correcta
    
    try:
        edad = int(input("Introduce tu edad: "))
        
        if edad < 0 or edad > 120:
            print("Error: La edad debe estar entre 0 y 120 años.")
        else:
            print(f"Edad válida: {edad} años")
            
    except ValueError:
        print("Error: Debes introducir un número entero.")
```

## EJERCICIO 12
### Genera una función que, al recibir una frase, devuelva una lista
### con la longitud de cada palabra. Usa la función map()

```python
def longitud_palabras(frase):
    # Devuelve una lista con la longitud de cada palabra usando map()
    
    palabras = frase.split()
    return list(map(lambda palabra: len(palabra), palabras))

# Prueba
frase = "Python es un lenguaje de progrmación"
print("Ejercicio 12:", longitud_palabras(frase))
```

## EJERCICIO 13
### Genera una función que, para un conjunto de caracteres, devuelva
### una lista de tuplas con cada letra en mayúsculas y minúsculas.
### Las letras no pueden estar repetidas. Usa la función map()



```python
def letras_mayusculas_minusculas(caracteres):
    # Devuelve una lista de tuplas con cada letra en mayúsculas y minúsculas
    caracteres_unicos = set(caracteres)
    
    return list(map(lambda letra: (letra.upper(), letra.lower()), caracteres_unicos))

# Prueba
caracteres = "aabbcc"
print("Ejercicio 13:", letras_mayusculas_minusculas(caracteres))
```

## EJERCICIO 14
### Crea una función que retorne las palabras de una lista que 
### comiencen con una letra en específico. Usa la función filter().

```python
def filtrar_por_inicial(palabras, letra_inicial):
    # Filtra palabras que comiencen con una letra específica
    
    return list(filter(lambda palabra: palabra.startswith(letra_inicial), palabras))

# Prueba
palabras = ["python", "progrmación", "java", "javascript", "php"]
print("Ejercicio 14:", filtrar_por_inicial(palabras, "p"))
```

## EJERCICIO 16
### Escribe una función que tome una cadena de texto y un número entero n
### como parámetros y devuelve una lista de todas las palabras que sean
### más largas que n. Usa la función filter().

```python
def palabras_mas_largas(texto, n):
    # Filtra palabras más largas que n caracteres
    
    palabras = texto.split()
    return list(filter(lambda palabra: len(palabra) > n, palabras))

# Prueba
texto = "Python es un lenguaje de progrmación muy poderoso"
print("Ejercicio 16:", palabras_mas_largas(texto, 5))
```
